```python
#Visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # pyplot is class in matplotlib
```

```python
#We create dataframe of house price
#we create a list of Area and Price of house
area=[600,800,1000,1200,1500,1800]
price=[1000000,1200000,1500000,1800000,2200000,2500000]
#Create dataframe from list 
df=pd.DataFrame(list(zip(area,price)), columns=['Area','Price']) # make new list they store value in new list then we will create df 
df
```

```python
#syntax of scatter plot : to show relation between input and output variable
#plt.scatter(X,Y) #where X=independent variable means input and Y : dependent
#means output variable
X=df["Area"]
Y=df["Price"]
plt.scatter(X,Y)
plt.show()
```

```python
#syntax of scatter plot : to show relation between input and output variable
#plt.scatter(X,Y) #where X=independent variable means input and Y : dependent
#means output variable
X=df["Area"]
Y=df["Price"]
plt.scatter(X,Y,c="red",s=70,marker="*") # to change dots color ,by default its blue /marker for symbol of dots/ s for size of data point
plt.xlabel("Area in sq. ft (Independent var)") # for x label
plt.ylabel("Price in lacs (dependent var)") # for y label name
plt.show()   #when we need only graph  not show address
```

```python
X=df["Area"]
Y=df["Price"]
plt.scatter(X,Y,c="red",s=70,marker="*") #c for color bydefault blue color
#s is size of data point which unit is pixel .pixel means smallest unit
#of graphics
plt.xticks(X) # to show exact value on point
plt.yticks(Y)
plt.xlabel("Area in Sq. Ft.")
plt.ylabel("Price in Lacs")
plt.show()
```

```python
#line graph : - use plot() inbuilt function of matplotlib.pyplot class 
X=df["Area"] #independent variable means input
Y=df["Price"]#dependent variable means output
plt.plot(X,Y,c="red",linewidth=5,linestyle="--") # to change line width give parameter  and line style
plt.xticks(X)
plt.yticks(Y)
plt.xlabel("Area in Sq. Ft.")
plt.ylabel("Price in Lacs")
plt.show()
```

```python
#line graph and scatter plot: -
#use plot() inbuilt function of matplotlib.pyplot class 
X=df["Area"] #independent variable means input
Y=df["Price"]#dependent variable means output
plt.scatter(X,Y,c="green",s=200,marker="*")
plt.plot(X,Y,c="blue",linewidth=3)
plt.xticks(X)
plt.yticks(Y)
plt.xlabel("Area in Sq. Ft.")
plt.ylabel("Price in Lacs")
plt.show()
```
