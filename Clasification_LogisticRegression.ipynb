```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#To load dataset mobile.csv
df=pd.read_csv("insurance.csv")
#to display first 5 records
df.head()
```

```python
#This dataset will be solved by  classification algorithm
#because we want to predict 1 or 0 means either customer claim for insurance or not

#basic steps  apply in dataset : -
#first check null values
df.isnull().sum()
```

```python
#Check how many rows and columns
df.shape
```

```python
#check datatype
df.info()
```

```python
#separate input and output variables from dataset 
X=df.drop("insuranceclaim",axis=1)  #input features
Y=df["insuranceclaim"] #output means target
```

```python
#Train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Use Logistic Regression algorithm because dataser is classification dataset ,so call 
#LogisticRegression inbuilt class similar as LinearRegression 
from sklearn.linear_model import LogisticRegression
```

```python
#create object of LogisticRegression similar as LinearRegression
lr=LogisticRegression() #here lr user defined object of LogisticRegression class
```

```python
#we train the model with the help of LogisticRegression , use inbuilt method fit() for train the model
#same as Linear regression (70% data)
lr.fit(X_train,Y_train)
```

```python
#testing the model with 30% data , we use inbuilt method predict() : formula Y_pred=1/(1+exp(-X))
#its gives value of probability (0-1) if prob value>0.5 then consider 1 otherwise=0
#ML always gives predicted o/p
Y_pred=lr.predict(X_test)
```

```python
#select Input X and output Y from dataset df
X=df.drop("insuranceclaim",axis=1)
Y=df["insuranceclaim"]
```

```python
#Train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Model train with the classification algorithm decisiontreeclassifier 
#DecisionTreeClassifier select input features from dataset with the help of gini index
#gini index find the impuries of each column
#find reduction = impurities of output column - impurities of each column
#select imp input column whose reduction is highest reduction
#find the impurities through GINI INDEX=1-P(Yes)^2-P(No)^2

from sklearn.tree import DecisionTreeClassifier  #DecisionTreeClassifier inbuilt class
```

```python
#create the object of DecisionTreeClassifier class
dt=DecisionTreeClassifier() # by default GINI Index
```

```python
#train the model then use fit() inbuilt method of DecisionTreeClassifier
dt.fit(X_train,Y_train) #train the model with 70% data
```

```python
#Test the model with 30% testing data use predict() inbuilt method
Y_pred=dt.predict(X_test)
```

```python
#check score means print classification_report 
print(classification_report(Y_test,Y_pred))
```

```python
#print confusion matrix
confusion_matrix(Y_test,Y_pred)
```

```python
#DecisionTree is perfect model in this dataset(insurance.csv). DT can automatic select the best 
#input features. For feature selection purpose , DecisionTree are always used . 

#Lets check out what are the important input features that are selected by the decisiontree gini
#index : -
#we use inbuilt method feature_importances_ of DecisionTreeClassifier class(show reduction 
#and select highest reduction)
dt.feature_importances_
```
