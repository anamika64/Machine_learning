```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create the use defined function
def sigmoid(X) : #sigmoid() user defined function and passing the parameter X (input)
    return (1/(1+np.exp(-X)))
```

```python
#call in main program
X=np.arange(-10,11) #start=-10  stop=11-1=10  step=+1
print(X)  #input
```

```python
#call function
Y=sigmoid(X) #call function
print(Y)
Y1=np.round(Y,2)
print(Y1)
```

```python
#Plot Value of X and Y with line (plot() inbuilt function )___ this give probability value
plt.figure(figsize=(25,25))
plt.plot(X,Y)
plt.title("Sigmoid function")
plt.xticks(X)
plt.yticks(Y)
plt.xlabel("Value of X input")
plt.ylabel("Value of Y output")
plt.show()
```
