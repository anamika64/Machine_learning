# 1.MinMaxScaler :

```python
#1.MinMaxScaler : -
#for example of minmaxscaler
#we create a one dimension numpy array A and B
A=np.array([10,20,30,40,50])
B=np.array([100,200,300,400,500])
```

```python
#1.MinMaxScaler : -
#for example of minmaxscaler
#we create a one dimension numpy array A and B
A=np.array([10,20,30,40,50])
B=np.array([100,200,300,400,500])
print(A)
```

```python
print(B)
```

```python
def minmax(arr): #passing arguments/parameters and return value
    arr_new=(arr-arr.min())/(arr.max()-arr.min())
    return arr_new
```

```python
#call function
minmax(A)
```
